## We will be building a movie recommender system and deploying it using Heroku

There are 2 types of recommender systems

1) Content Based - Suggests items based on content consumed. E.g: If you are listening romantic songs, you will be suggested romantic songs only
    
2) Collaborative Filtering - It recommends based on behaviours of users. Suppose A and B are 2 friends having similar taste 
in movies, then if A likes 3 Idiots, then the system will also try to recommend the same to B. This is also seen in 
social media. E.g - If a connection/friend on linkedin likes / reacts to a post, then as you and your friend are very similar,
Linkedin shows their reacted posts on your feed too

## Thus we need to extract value of the "name" key

```python
import ast
def convert(obj):
    
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

## The format is same for the keywords column so we apply the same convert function

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

## Now there are many actors in a single movie, but we will take only the top 3 for consideration

```python
import ast
def convert_cast(obj):
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if(counter!=3):
            L.append(i['name'])
            counter+=1
        else:
            break
        
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

## Now we will get the director's name for each movie. We are not bothered about other roles

```python
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
        
        
    return L
```

```python
movies['crew'].apply(get_director)
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
## NOW WE SEE THAT OVERVIEW IS A STRING WHILE OUR OTHER FEATURES PREPROCESSED TILL NOW ARE LISTS.

movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

## Now we can proceed to add all the other 4 columns' info into the overview list. But there is a problem

## We see there are spaces in between words like Sam Worthington. While building our system, different entities/tags  will be created for Sam and Worthington. We also see there is another guy Sam Mendes. So for him also, 2 tags will be created - Sam and Mendes. So it might happen that if we are trying to search movies of Sam Worthington , we may end up getting recommended those of Sam Mendes So to avoid this we will group those words into a single word so that single and unique entities are created

```python
movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

## Text Vectorization

Now to determine the closeness of each movie or how much they have in common, we cannot the relationship using text only. Text needs to be converted into vectors and then the closes vectors are chosen and the corresponding movies are recommended. There are many techniques such as Bag Of Words, TF-IDF, Word2Vec. We will be using BOW.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

We are not considering stopwords like a, an , and etc which have no role in contributing meaning to a sentence

## But we see that there is one problem that similar words like action and actions are acting as different features. So we can apply stemming to the tags column. 

## E.g: If we have a list before stemming like ['loving', 'loved', 'love'], then the list after stemming will be ['love', 'love', 'love']

Importing NLTK Package

```python
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
## To get the no of unique words we can use
cv.get_feature_names()
```

## Now we will calculate the distance between 2 vectors(movies). We will not use euclidean distance , instead we will use cosine distance or the angle between them. This is because euclidean distance( tip to tip distance) fails in case of higher dimensions

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0]
```

The above row (1st row) is the similarity of the 1st movie with other movies

##  Now we will define a function which will recommend 5 most closest movies. 

```python
def recommend(movie):
    
    movie_index= new_df[new_df['title']==movie].index[0]  # TO FIND INDEX
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        
        print(new_df.iloc[i[0]].title)
    return
```

## One problem encountered is that while searching for top 5 closest vectors, we sort the vectors . While sorting the vectors, the indexes of the original vectors will be lost . So we use enumerate to create a tuple which will keep track of the indexes as well.

```python
recommend('Batman Begins')
```
